```python
import math
import matplotlib.pyplot as plt
import numpy as np

# Datos
xi = [1, 2, 3, 4, 5, 6, 7, 8, 9]
yi = [1, 1.5, 2, 3, 4, 5, 8, 10, 13]

# Suma (xi * yi)
multi_xi_yi = [x * y for x, y in zip(xi, yi)]
suma_multi_xi_yi = sum(multi_xi_yi)

# Suma xi
Sum_xi = sum(xi)

# Suma yi
Sum_yi = sum(yi)

# Xi cuadrado
cuadrado_xi = [x**2 for x in xi]
suma_cuadrado_xi = sum(cuadrado_xi)

# a1 (pendiente) y a0 (intercepto)
n = len(xi)
a1 = ((n * suma_multi_xi_yi) - (Sum_xi * Sum_yi)) / ((n * suma_cuadrado_xi) - (Sum_xi**2))
a0 = (Sum_yi - a1 * Sum_xi) / n

print("Modelo: y =", round(a1, 5), "x +", round(a0, 5))

# Método A) Cálculo de Sr y Syx
Sr = sum([(y - a0 - a1 * x)**2 for x, y in zip(xi, yi)])
Syx = math.sqrt(Sr / (n - 2))

print("Sr:", round(Sr, 5))
print("Syx:", round(Syx, 5))

# Método B) Cálculo de Sy
prom = sum(yi) / len(yi)
Sy = math.sqrt(sum([(y - prom)**2 for y in yi]) / (n - 1))

print("Sy:", round(Sy, 5))

# Graficar los datos y la recta ajustada
x_vals = np.linspace(min(xi), max(xi), 100)
y_vals = a0 + a1 * x_vals

plt.scatter(xi, yi, color='blue', label="Datos")
plt.plot(x_vals, y_vals, color='red', label=f"Recta ajustada: y = {round(a1, 2)}x + {round(a0, 2)}")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regresión por Mínimos Cuadrados")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Newton-Raphson usando P0=0.3, encontrar P3

# Definir la función f(x) = 7sin(x)e^(-x) - 1
def f(x):
    return 7 * np.sin(x) * np.exp(-x) - 1

# Definir la derivada de f(x)
def df(x):
    return 7 * np.exp(-x) * (np.cos(x) - np.sin(x))

# Semilla inicial
P0 = 0.3

# Aplicar el método de Newton-Raphson
for i in range(3):
    fn = f(P0)   # f(P0)
    dfn = df(P0) # f'(P0)

    # Fórmula del método de Newton-Raphson
    Pn = P0 - fn / dfn

    # Calcular el error absoluto
    error_abs = abs(Pn - P0)
    print(f"Iteración {i+1}: Pn = {Pn:.6f}, Error absoluto = {error_abs:.6f}")

    # If para detener las iteraciones si el error es suficientemente pequeño
    if error_abs < 1e-6:
        break

    # Se actualiza para ir iterando
    P0 = Pn

# Rango de valores para x
x = np.linspace(0, 2, 400)
y = f(x)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = 7\sin(x)e^{-x} - 1$', color='b')
plt.axhline(0, color='black', linewidth=1)  # Eje X
plt.axvline(0, color='black', linewidth=1)  # Eje Y
plt.scatter(Pn, f(Pn), color='r', label="Raíz")  # Punto final
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Método de Newton-Raphson para $f(x) = 7\sin(x)e^{-x} - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Diferencias divididas

import numpy as np
import matplotlib.pyplot as plt

x = [0, 1.8, 5, 6, 8.2]
y = [26, 16.415, 5.375, 3.5, 2.015]

# y[i+1], x[i+1] va incrementando respecto a la longitud de la lista x, que son 5 valores
dif_1 = [(y[i+1] - y[i]) / (x[i+1] - x[i]) for i in range(len(x)-1)]
print(dif_1)

dif_2 = [(dif_1[i+1] - dif_1[i]) / (x[i+2] - x[i]) for i in range(len(dif_1)-1)]
print(dif_2)

dif_3 = [(dif_2[i+1] - dif_2[i]) / (x[i+3] - x[i]) for i in range(len(dif_2)-1)]
print(dif_3)

dif_4 = [(dif_3[i+1] - dif_3[i]) / (x[i+4] - x[i]) for i in range(len(dif_3)-1)]
print(dif_4)

coef_0 = y[0]  # f[x0]
coef_1 = dif_1[0]  # f[x0, x1]
coef_2 = dif_2[0]  # f[x0, x1, x2]
coef_3 = dif_3[0]  # f[x0, x1, x2, x3]
coef_4 = dif_4[0]  # f[x0, x1, x2, x3, x4]

n = 3.5
P = (coef_0 +
    coef_1 * (n - x[0]) +
    coef_2 * (n - x[0]) * (n - x[1]) +
    coef_3 * (n - x[0]) * (n - x[1]) * (n - x[2]) +
    coef_4 * (n - x[0]) * (n - x[1]) * (n - x[2]) * (n - x[3]))

# Imprimir el valor del polinomio en n = 1.5
print(f"P(3.5) = {P}")

# Graficar el polinomio sobre los datos originales
x_interp = np.linspace(min(x), max(x), 100)
y_interp = [coef_0 +
    coef_1 * (xi - x[0]) +
    coef_2 * (xi - x[0]) * (xi - x[1]) +
    coef_3 * (xi - x[0]) * (xi - x[1]) * (xi - x[2]) +
    coef_4 * (xi - x[0]) * (xi - x[1]) * (xi - x[2]) * (xi - x[3])
    for xi in x_interp]

plt.scatter(x, y, color='red', label="Datos")
plt.plot(x_interp, y_interp, color='blue', label="Polinomio de Newton")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.title("Interpolación de Newton")
plt.show()
```

```python
# Método de Newton-Raphson para el polinomio dado
# f(x) = 0.0074x^4 - 0.284x^3 + 3.355x^2 - 12.183x + 5
# Semilla inicial x0 = 16.15

# Definir la función y su derivada
def f(x):
    return 0.0074*x**4 - 0.284*x**3 + 3.355*x**2 - 12.183*x + 5

def df(x):
    return 4*0.0074*x**3 - 3*0.284*x**2 + 2*3.355*x - 12.183

# Semilla inicial
P0 = 16.5

# Aplicación del método de Newton-Raphson
for i in range(10):
    fn = f(P0)  # f(P0)
    dfn = df(P0)  # f'(P0)

    # Evitar división por cero
    if dfn == 0:
        print("La derivada es cero, el método falla en esta iteración.")
        break

    # Fórmula de Newton-Raphson
    Pn = P0 - fn / dfn

    # Calcular el error absoluto
    error_abs = abs(Pn - P0)
    print(f"Iteración {i+1}: Pn = {Pn:.6f}, Error absoluto = {error_abs:.6f}")

    # Condición de parada
    if error_abs < 1e-6:
        break

    # Actualizar P0
    P0 = Pn

# Gráfica de la función
x = np.linspace(-10, 30, 500)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = 0.0074x^4 - 0.284x^3 + 3.355x^2 - 12.183x + 5$', color='b')
plt.axhline(0, color='black', linewidth=1)  # Eje X
plt.axvline(0, color='black', linewidth=1)  # Eje Y
plt.scatter(Pn, f(Pn), color='r', label="Raíz encontrada")  # Punto final
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Método de Newton-Raphson aplicado a $f(x)$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
R = 0.518  # kJ/kg·K
Tc = 191  # Temperatura crítica (K)
Pc = 4600  # Presión crítica (kPa)

# Cálculo de los parámetros a y b
a = 0.427 * ((R**2 * Tc**2.5) / Pc)
b = 0.0866 * (R * Tc / Pc)

# Datos del problema
T = -40 + 273.15  # Conversión de °C a K
P = 65000  # Presión absoluta (kPa)
V_tanque = 3  # Volumen total del tanque (m³)

# Definir la ecuación de Redlich-Kwong como función de v
def f(v):
    return ((R * T) / (v - b)) - (a / ((v * (v + b)) * (T**0.5))) - P

# Semillas iniciales (valores razonables para el volumen específico)
P0 = 0.0005  # Suposición inicial para el volumen específico
P1 = 0.001   # Otro valor inicial

# Parámetros del método de la secante
tolerance = 1e-6  # Criterio de convergencia
max_iter = 100  # Número máximo de iteraciones

# Método de la secante
for i in range(max_iter):
    fn_1 = f(P0)  # f(P0)
    fn = f(P1)    # f(P1)

    # Evitar división por cero
    if abs(fn - fn_1) < 1e-12:
        break

    # Fórmula del método de la secante
    Pn = P1 - (fn * (P1 - P0)) / (fn - fn_1)

    # Calcular el error absoluto
    error_abs = abs(Pn - P1)

    # If para detener las iteraciones si el error es menor que la tolerancia
    if error_abs < tolerance:
        break

    # Se actualizan los valores
    P0, P1 = P1, Pn  

# Volumen específico encontrado
v_sol = Pn

# Cálculo de la masa de metano en el tanque: masa = V_tanque / v_específico
masa_metano = V_tanque / v_sol

# Mostrar resultados
print(f"Volumen específico encontrado: {v_sol:.6f} m³/kg")
print(f"Masa de metano contenida en el tanque: {masa_metano:.2f} kg")

# Graficar la función para visualizar su comportamiento
x = np.linspace(0.0001, 0.01, 400)
y = [f(val) for val in x]

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(v)$', color='b')
plt.axhline(0, color='black', linewidth=1, linestyle='--')  # Eje X
plt.axvline(v_sol, color='red', linestyle='--', label=f'v ≈ {v_sol:.6f} m³/kg')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Gráfica de la función de Redlich-Kwong")
plt.xlabel("Volumen específico (m³/kg)")
plt.ylabel("f(v)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de la secante para f(x) = 7sin(x)e^(-x) - 1

# Definir la función
def f(x):
    return 7 * np.sin(x) * np.exp(-x) - 1

# Semillas iniciales
P0 = 0.5
P1 = 0.4

# Aplicar el método de la secante
for i in range(5):
    fn_1 = f(P0)  # f(P0)
    fn = f(P1)    # f(P1)

    # Fórmula del método de la secante
    Pn = P1 - (fn * (P1 - P0)) / (fn - fn_1)

    # Calcular el error absoluto
    error_abs = abs(Pn - P1)
    print(f"Iteración {i+1}: Pn = {Pn:.6f}, Error absoluto = {error_abs:.6f}")
    
    # If para detener las iteraciones si el error es suficientemente pequeño
    if error_abs < 1e-6:
        break

    # Se actualizan los valores en la iteración
    P0, P1 = P1, Pn  

# Rango de valores para x
x = np.linspace(0, 2, 400)
y = f(x)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = 7\sin(x)e^{-x} - 1$', color='b')
plt.axhline(0, color='black', linewidth=1)  # Eje X
plt.axvline(0, color='black', linewidth=1)  # Eje Y
plt.scatter(Pn, f(Pn), color='r', label="Raíz")  # Punto final
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title("Método de la Secante para $f(x) = 7\sin(x)e^{-x} - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
